```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline


import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import datasets, ensemble
from sklearn.metrics import mean_squared_error
from sklearn.inspection import permutation_importance

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestRegressor

import sklearn

from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import keras


import xgboost as xgb
```

```python
df = pd.read_csv('train.csv', parse_dates=[1], dayfirst=True)
df['year'] = pd.to_datetime(df.Date).dt.year
df['month'] = pd.to_datetime(df.Date).dt.month
df['day'] = pd.to_datetime(df.Date).dt.day
df['time'] = pd.to_datetime(df[['year', 'month', 'day', 'Hour']])
df = df.set_index('time')
df.drop(['year', 'month', 'day'], axis=1, inplace=True)
df['weekday'] = pd.to_datetime(df.Date).dt.weekday
df.head(2)
```

```python
df['work_day'] = np.where((df['weekday'] == 5) | (df['weekday'] == 6) , False, True)
df.loc[df.Holiday == 'Holiday', 'work_day'] = False
df[df['Holiday']=='Holiday']
```

```python
#Heat index 
df['Temp_F'] = 1.8*df['Temperature'] + 32
T = df['Temp_F']
RH = df['Humidity']
df['heat_index'] = -42.379 + 2.04901523*T + 10.14333127*RH - .22475541*T*RH - .00683783*T*T - .05481717*RH*RH + .00122874*T*T*RH + .00085282*T*RH*RH - .00000199*T*T*RH*RH
```

```python
#lasso 
from sklearn import linear_model
clf = linear_model.Lasso(alpha=0.05)
X = df[[#'Temperature',
    #'Humidity',
        #'Wind speed', 
        #'Visibility',
       'Dew point temperature',
        'Solar Radiation',
        'Rainfall',
       'Snowfall' ,
        #'heat_index',
        #'Temp_F'
       ]]
y = df['y']
clf.fit(X, y)


df['y_lasso'] = clf.predict(X)
```

```python
df['Visibility'] = np.log(df['Visibility'])
df.head(2)
```

```python
a=3   #best 3
b=5   #best 5
c=24
df['Rainfall_MA'] = df['Rainfall'].rolling(window=a, min_periods=1).mean()
df['Snowfall_MA'] = df['Snowfall'].rolling(window=b, min_periods=1).mean()
df['Temperature_MA'] = df['Temperature'].rolling(window=c, min_periods=1).mean()
```

```python
df["Functioning Day"] = df["Functioning Day"].replace({'Yes': 1, 'No': 0})
df["Holiday"] = df["Holiday"].replace({'No Holiday': 1 , 'Holiday': 0})
df["Seasons"] = df["Seasons"].replace({'Winter': 1, 'Spring': 2,"Summer": 3, "Autumn": 4})

```

```python
#New Feature
df['combo'] = (df['Solar Radiation'] - df['Rainfall'] - df['Snowfall']) / (df['Solar Radiation'] + df['Wind speed'])
df['combo'].value_counts()
```

```python
#PCA for Solar Radiatio, raining and snowfall:
from sklearn import decomposition

SRS = df[['Solar Radiation', 'Rainfall', 'Snowfall' ]].T

pca = decomposition.PCA(n_components=3)
pca.fit(SRS)
print(pca.components_.T.shape)
new_comps = pd.DataFrame(pca.components_.T, columns=['comp1', 'comp2', 'comp3'])
new_comps.describe().T
```

```python
df['comp1'] = pca.components_[0]
df['comp2'] = pca.components_[1]
df['comp3'] = pca.components_[2]
```

```python
feats = [#'Date', 
                'Hour',
                'Temperature',
                'Humidity',
                'Wind speed',
                'Visibility',
                #'Dew point temperature',
                'Solar Radiation',
                'Rainfall',
                'Snowfall',
                #'Seasons',
                'Holiday',
                'Functioning Day',
                #'weekday',
                #'Day sin',
                #'Day cos',
                'week sin',
                'week cos',
                'Year sin',
                'Year cos',
                #'comp1',
                #'comp2',
                #'comp3',
                #'tsne-2d-one',
                #'tsne-2d-two'
                'Rainfall_MA',
                'Snowfall_MA',
                #'Temperature_MA',
                #'Wind speed_MA'
                #'work_day',
                #'heat_index',
                #'Temp_F',
                #'y_lasso'
                
                
]

xtrain_xgb = df[feats]
ytrain_xgb = df['y']
ytrain_xgb_log = np.log(df['y']).replace(np.NINF, 0)

xtest_xgb = df_test[feats]
ytest_xgb = df_test['count']
ytest_xgb_log = np.log(ytest_xgb).replace(np.NINF, 0)

import xgboost as xgb

xgbr = xgb.XGBRegressor(max_depth=50, n_estimators=50, gamma=0.57,min_child_weight=90,
                        max_delta_step=3, learning_rate=0.2, random_state=42,verbosity=0, alpha=0.02 ) 
xgbr.fit(xtrain_xgb, ytrain_xgb_log)

ypred_log = xgbr.predict(xtest_xgb)
ypred = np.exp(ypred_log)
ypred[ytest_xgb==0] = 0
print((ypred<0).sum())
ypred[ypred<0] =0

msle = sklearn.metrics.mean_squared_log_error(ytest_xgb, ypred)
rmsle = np.sqrt(msle)
print("RMSLE = ",rmsle)

mse = sklearn.metrics.mean_squared_error(ytest_xgb, ypred)
rmse = np.sqrt(mse)
print("RMSE = ",rmse)
```

```python
#neural network
model = Sequential()
model.add(Dense(256, input_dim=16, activation='relu'))
model.add(Dense(1))
# Compile model
model.compile(loss="msle", optimizer='adam', metrics=['mean_squared_logarithmic_error'])
# Fit the model
history = model.fit(xtrain_xgb, ytrain_xgb_log, validation_split=0.20, epochs=100, batch_size=10, verbose=1)

ypred_log = model.predict(xtest_xgb)
ypred = np.exp(ypred_log)
ypred[ytest_xgb==0] = 0
print((ypred<0).sum())
ypred[ypred<0] =0

msle = sklearn.metrics.mean_squared_log_error(ytest_xgb, ypred)
rmsle = np.sqrt(msle)
print("RMSLE = ",rmsle)

mse = sklearn.metrics.mean_squared_error(ytest_xgb, ypred)
rmse = np.sqrt(mse)
print("RMSE = ",rmse)
```

```python
#SVM
from sklearn.svm import SVR
svm = SVR(C=1.0, epsilon=0.2)
svm.fit(xtrain_xgb, ytrain_xgb_log)

ypred_log = svm.predict(xtest_xgb)
ypred = np.exp(ypred_log)
ypred[ytest_xgb==0] = 0
print((ypred<0).sum())

msle = sklearn.metrics.mean_squared_log_error(ytest_xgb, ypred)
rmsle = np.sqrt(msle)
print("RMSLE = ",rmsle)

mse = sklearn.metrics.mean_squared_error(ytest_xgb, ypred)
rmse = np.sqrt(mse)
print("RMSE = ",rmse)
```
