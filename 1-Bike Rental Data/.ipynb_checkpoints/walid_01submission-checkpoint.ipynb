# Data Cleanining and validation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline


import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import datasets, ensemble
from sklearn.metrics import mean_squared_error
from sklearn.inspection import permutation_importance

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestRegressor

import sklearn

import xgboost as xgb
```

```python
df = pd.read_csv('Data/train.csv', parse_dates=[1], dayfirst=True)
df['year'] = pd.to_datetime(df.Date).dt.year
df['month'] = pd.to_datetime(df.Date).dt.month
df['day'] = pd.to_datetime(df.Date).dt.day
df['time'] = pd.to_datetime(df[['year', 'month', 'day', 'Hour']])
df = df.set_index('time')
df.drop(['year', 'month', 'day'], axis=1, inplace=True)
df['weekday'] = pd.to_datetime(df.Date).dt.weekday
df.head(2)
```

```python
df["Functioning Day"] = df["Functioning Day"].replace({'Yes': 1, 'No': 0})
df["Holiday"] = df["Holiday"].replace({'No Holiday': 1 , 'Holiday': 0})
```

```python
time = pd.DataFrame(pd.date_range("2017-12-01", "2018-11-30", freq="1H"), columns=['time'])
df_full_range = time.merge(df, how='left', on='time')
df_full_range = df_full_range.set_index('time')


ax = plt.figure(figsize=(18,5))
df_full_range['y'].loc[df_full_range.index.month==(12)].plot()
for month in range(11):
    plt.figure(figsize=(18,5))
    df_full_range['y'].loc[df_full_range.index.month==(month+1)].plot()
    
```

```python
timestamp_s = df.index.map(pd.Timestamp.timestamp)

print(timestamp_s)
day = 24*60*60
week = 7 * day
year = (365.2425)*day

df['Day sin'] = np.sin(timestamp_s * (2 * np.pi / day))
df['Day cos'] = np.cos(timestamp_s * (2 * np.pi / day))
df['week sin'] = np.sin(timestamp_s * (2 * np.pi / week))
df['week cos'] = np.cos(timestamp_s * (2 * np.pi / week))
df['Year sin'] = np.sin(timestamp_s * (2 * np.pi / year))
df['Year cos'] = np.cos(timestamp_s * (2 * np.pi / year))
```

```python
feats = [#'Date', 
                'Hour',
                'Temperature',
                'Humidity',
                'Wind speed',
                'Visibility',
                #'Dew point temperature',
                'Solar Radiation',
                'Rainfall',
                'Snowfall',
                #'Seasons',
                'Holiday',
                'Functioning Day',
                #'weekday',
                #'Day sin',
                #'Day cos',
                'week sin',
                'week cos',
                'Year sin',
                'Year cos'
]

xtrain_xgb = df[feats]
ytrain_xgb = df['y']
ytrain_xgb_log = np.log(df['y']).replace(np.NINF, 0)

import xgboost as xgb

xgbr = xgb.XGBRegressor(max_depth=50, n_estimators=50, learning_rate=0.2, random_state=42,verbosity=0) 
xgbr.fit(xtrain_xgb, ytrain_xgb_log)



```

```python
#Test Data
df_test = pd.read_csv('last_10_days.csv')

df_test['year'] = pd.to_datetime(df_test.Date).dt.year
df_test['month'] = pd.to_datetime(df_test.Date).dt.month
df_test['day'] = pd.to_datetime(df_test.Date).dt.day
df_test['time'] = pd.to_datetime(df_test[['year', 'month', 'day', 'Hour']])
df_test = df_test.set_index('time')
df_test.drop(['year', 'month', 'day'], axis=1, inplace=True)
df_test['weekday'] = pd.to_datetime(df_test.Date).dt.weekday
df_test.rename(columns={'Rented Bike Count': 'count', 'Temperature(°C)':'Temperature(C)'}, inplace=True)
df_test.head()
```

```python
df_test = df_test[['Date','count','Hour', 'Temperature',
       'Humidity', 'Wind speed', 'Visibility',
       'Dew point temperature', 'Solar Radiation', 'Rainfall',
       'Snowfall', 'Holiday', 'Functioning Day', 'weekday']]
df_test.head(2)
```

```python
timestamp_s = df_test.index.map(pd.Timestamp.timestamp)


day = 24*60*60
week = 7 * day
year = (365.2425)*day

df_test['Day sin'] = np.sin(timestamp_s * (2 * np.pi / day))
df_test['Day cos'] = np.cos(timestamp_s * (2 * np.pi / day))
df_test['week sin'] = np.sin(timestamp_s * (2 * np.pi / week))
df_test['week cos'] = np.cos(timestamp_s * (2 * np.pi / week))
df_test['Year sin'] = np.sin(timestamp_s * (2 * np.pi / year))
df_test['Year cos'] = np.cos(timestamp_s * (2 * np.pi / year))
```

```python
df_test.columns
```

```python
xtest_xgb = df_test[feats]
ytest_xgb = df_test['count']

ytest_xgb_log = np.log(ytest_xgb).replace(np.NINF, 0)
ytest_xgb_log.value_counts()
```

```python

ypred_log = xgbr.predict(xtest_xgb)
ypred = np.exp(ypred_log)
ypred[ytest_xgb==0] = 0
ypred[ypred<0] =0

msle = sklearn.metrics.mean_squared_log_error(ytest_xgb, ypred)
rmsle = np.sqrt(msle)
print("RMSLE = ",rmsle)

mse = sklearn.metrics.mean_squared_error(ytest_xgb, ypred)
rmse = np.sqrt(mse)
print("RMSE = ",rmse)
```
